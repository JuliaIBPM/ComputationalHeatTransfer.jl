Churchill friction correlation

\begin{equation}
f / 8=\left[\left(\frac{8}{\operatorname{Re}}\right)^{12}+\frac{1}{\left(\Theta_{1}+\Theta_{2}\right)^{1.5}}\right]^{\frac{1}{12}}
\end{equation}

where

\begin{equation}
\begin{array}{l}
\Theta_{1}=\left[-2.457 \ln \left(\left(\frac{7}{\operatorname{Re}}\right)^{0.9}+0.27 \frac{\varepsilon}{D}\right)\right]^{16} \\
\Theta_{2}=\left(\frac{37530}{\operatorname{Re}}\right)^{16}
\end{array}
\end{equation}

```julia
Re = 1e3
系 = 0.001
```

```julia
function f_churchill(Re,系)
    1 = (-2.457*log((7/Re)^(0.9)  +  0.27 * 系))^16
    2 = (37530/Re)^16
    f=8*((8/Re)^12+(1/(1+2)^1.5))^(1/12)
    
    f
end
```

### Darcy friction coefficient v.s. Reynolds number

```julia
Re_list = 1e3:10:1e4
系 = 0.001
fhist = f_churchill.(Re_list);
plot(Re_list,fhist,label=string("Churchill 系/D = ", 系))
plot!(Re_list,64 ./ Re_list,label="laminar",xlabel="Re",ylabel="f")
```
